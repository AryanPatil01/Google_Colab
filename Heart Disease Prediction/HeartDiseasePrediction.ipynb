```python
import pandas as pd
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
heart_df = pd.read_csv('D:\Google_Colab\Heart Disease Prediction\heart_disease_data.csv')
```

```python
x = heart_df.drop(columns = 'target')
y = heart_df['target']
```

```python
x
```

```python
y


    
```

```python
x_train , x_test , y_train , y_test  = train_test_split( x , y , test_size=0.2 , random_state=1 ,stratify=y)
    
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
print(y.shape , y_train.shape , y_test.shape)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train , y_train)
```

```python
x_train_prediction = model.predict(x_train)
x_train_accuracy = accuracy_score(x_train_prediction , y_train)
```

```python
print("training_accuracy : " , x_train_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_prediction , y_test )
```

```python
print("testing_accuracy : " , x_test_accuracy)
```

```python
final_patient_data = [[55,0,0,128,205,0,2,130,1,2,1,1,3,]]
standardised_final_patient_data = scaler.transform(final_patient_data)
```

```python
evaluation_final_predict = model.predict(standardised_final_patient_data)
evaluation_final_probablity = model.predict_proba(standardised_final_patient_data)

```

```python
print("evaluation_acc_final : " , evaluation_final_predict)
```

```python
prone_to_heart_disease = [1]
Not_prone_to_heart_disease = [0]

if evaluation_final_predict == Not_prone_to_heart_disease :
    print("Patient is not prone to Heart diseases")
else:
     print("Patient is prone to Heart diseases")
```

```python
print("Heart disease probability:", evaluation_final_probablity[0][1]*100 )
```
